## Eric's Task

Find all the entries that belong to the same project.

Then, assign tags to each entry.

Given a input strings of:

```
sebsnyk/juliet-test-suite-csharp(main):src/testcases/CWE191_Integer_Underflow/s03/CWE191_Integer_Underflow_s03.csproj
sebsnyk/juliet-test-suite-csharp(main):src/testcases/CWE191_Integer_Underflow/s04/CWE191_Integer_Underflow_s04.csproj
```

We extract the project grouping ID `sebsnyk/juliet-test-suite-csharp`.

Note how `(main)` is being stripped away.

Now we can build a table of specific projects belonging to it and perform the tagging work.

````python
import snyk
import pandas as pd
````

````python
client = snyk.SnykClient("5d778343-78e7-4c73-9ba6-55af47a2f6af")
org = [org for org in client.organizations.all() if org.slug == 'e-corp'][0]
projects = org.projects.all()
````

````python
grouped = {}

for p in projects:
    key = p.remoteRepoUrl
        
    if key != None:
        if not key in grouped:
            grouped[key] = []
        grouped[key].append(p)
        next
    
    
    key = p.name.split(':')[0]
    if not key in grouped:
        grouped[key] = []

    grouped[key].append(p)
````

````python
of_interest = [key for key in grouped.keys() if key.startswith('sebsnyk/java-goof')][0]
df = pd.DataFrame(grouped[of_interest])
df
````

````python
print ('dependencies across all: {}'.format(df['totalDependencies'].sum()))

print ('dependencies across all: {}'.format(df['totalDependencies'].describe()))
````
